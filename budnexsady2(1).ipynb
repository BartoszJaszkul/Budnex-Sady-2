```python

# load the projectpro webpage content

r = requests.get('https://www.budnex.pl/gorzow_wlkp/mieszkania/sady_ii_bud2.html')



# convert to beautiful soup

soup = bs4(r.content)



# printing our web page

print(soup.prettify())
```

```python
table = soup.select('table#myTable-1')[0]
columns = table.find('thead').find_all('th')
df_miesz = pd.read_html(str(table))[0]
```

```python
table = soup.select('table#myTable-2')[0]
columns = table.find('thead').find_all('th')
df_garaze = pd.read_html(str(table))[0]
```

```python
table = soup.select('table#myTable-3')[0]
columns = table.find('thead').find_all('th')
df_piwnice = pd.read_html(str(table))[0]
```

```python
print(df_miesz.columns)
df_miesz.columns = ['Klatka', 'Pietro', 'Nr lokalu', 'Powierzchnia', 'Liczba pokoi', 'Balkon', 'Cena za m.kw', 'Status', 'Karta', 'Rezerwacja']
df_garaze.columns = ['Klatka', 'Pietro', 'Nr lokalu', 'Powierzchnia', 'Liczba pokoi', 'Balkon', 'Cena za m.kw', 'Status', 'Karta', 'Rezerwacja']
df_piwnice.columns = ['Klatka', 'Pietro', 'Nr lokalu', 'Powierzchnia', 'Liczba pokoi', 'Balkon', 'Cena za m.kw', 'Status', 'Karta', 'Rezerwacja']
print(df_miesz.columns)
print(df_garaze.columns)
print(df_piwnice.columns)
```

```python
df_miesz['Powierzchnia'] = df_miesz['Powierzchnia']/100
df_miesz['Q-ty'] = 1
df_miesz
```

```python
df_garaze['Powierzchnia'] = df_garaze['Powierzchnia']/100
df_garaze['Q-ty'] = 1
df_garaze
```

```python
df_piwnice['Powierzchnia'] = df_piwnice['Powierzchnia']/100
df_piwnice['Q-ty'] = 1
df_piwnice
```

```python
tabela_miesz = pd.pivot_table(df_miesz, values='Q-ty', index=['Liczba pokoi'], columns=['Status'], aggfunc=np.sum, fill_value=0)
tabela_miesz
```

```python
df_garaze['quantile'] = pd.qcut(df_garaze['Powierzchnia'], 4, labels=False)
df_garaze
tabela_garaz = pd.pivot_table(df_garaze, values='Q-ty', index=['quantile'], columns=['Status'], aggfunc=np.sum, fill_value=0)
tabela_garaz
```
